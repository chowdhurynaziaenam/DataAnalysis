```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# defining colors for each species
colors = {'Iris-setosa':'red', 'Iris-versicolor':'green', 'Iris-virginica':'blue'}
sns.countplot(x='Species', data=df, palette=colors, hue='Species', legend=False)
plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt


fig, axes = plt.subplots(2, 2, figsize=(10,15))

axes[0,0].set_title("Sepal Length")
axes[0,0].hist(df['SepalLengthCm'], bins=7)

axes[0,1].set_title("Sepal Width")
axes[0,1].hist(df['SepalWidthCm'], bins=5);

axes[1,0].set_title("Petal Length")
axes[1,0].hist(df['PetalLengthCm'], bins=6);

axes[1,1].set_title("Petal Width")
axes[1,1].hist(df['PetalWidthCm'], bins=6);
```

```python
#comparing petal length and petal width
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt


sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df, )

# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```


From the above plot, it can be said that

1. Versicolor Species lies in the middle of other two species both in terms of petal length and petal width.
2. Species Virginica has the largest of petal lengths and widths.
3. Species Setosa has relatively smaller petal lengths and widths.

```python
#comparing sepal length and sepal width
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',hue='Species', data=df, )

# placing legend outside figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

From the above plot, it can be said that

1. Species Virginica has comparatively larger sepal lengths but smaller sepal widths.
2. Versicolor Species lies in the middle of other two species in terms of both sepal length and width.
3. Species Setosa has smaller sepal lengths but relatively larger sepal widths.

```python
import numpy as np

# Load the dataset
df = pd.read_csv('Iris.csv')

# IQR
Q1 = np.percentile(df['SepalWidthCm'], 25,interpolation = 'midpoint')

Q3 = np.percentile(df['SepalWidthCm'], 75,interpolation = 'midpoint')
IQR = Q3 - Q1

print("Previous Shape: ", df.shape)

# Upper bound
upperbound = np.where(df['SepalWidthCm'] >= (Q3+1.5*IQR))

# Lower bound
lowerbound = np.where(df['SepalWidthCm'] <= (Q1-1.5*IQR))

# Removing the Outliers
df.drop(upperbound[0], inplace = True)
df.drop(lowerbound[0], inplace = True)

print("New Shape: ", df.shape)

sns.boxplot(x='SepalWidthCm', data=df)
plt.show()
```
