```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Iris.csv')

# Selecting only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Store outliers before capping
outliers_dict = {}

# Iterate over each numeric column and apply IQR method
for col in numeric_cols:
    Q1 = np.percentile(df[col], 25, interpolation='midpoint')
    Q3 = np.percentile(df[col], 75, interpolation='midpoint')
    IQR = Q3 - Q1

    # Define upper and lower bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Find outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][['Id', col]]

    # Store outliers for this column
    if not outliers.empty:
        outliers_dict[col] = outliers

    # Capping outliers instead of removing them
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])

# Print all detected outliers
print("\n **Outliers Detected in Each Column:**")
for col, outliers in outliers_dict.items():
    print(f"\n**{col} Outliers:**\n", outliers)

# Show boxplots for all numeric columns after capping
new_data = df.melt(id_vars=['Id'], value_vars=numeric_cols, var_name='Feature', value_name='Value')

plt.figure(figsize=(12, 6))
sns.boxplot(x='Feature', y='Value', data=new_data)
plt.xticks(rotation=15)
plt.title("Boxplots After Capping Outliers")
plt.show()

print("New Shape After Capping: ", df.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df.head(10)
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=['Species'])
Y=df['Species']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

#Apply MinMaxScaler to the data

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)


# Initialize classifiers
classifiers = {
    'SVM': SVC(probability=True),
    'Logistic Regression': LogisticRegression(),
    'K-NN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier()
}

# Store results
results = []

# Train and evaluate each classifier
for name, clf in classifiers.items():
    clf.fit(X_train_scaled, y_train)
    training_prediction = clf.predict(X_train_scaled)
    training_accuracy = accuracy_score(y_train, training_prediction)
    y_prediction = clf.predict(X_test_scaled)


    # Calculate accuracy and store it in a variable
    accuracy = accuracy_score(y_test, y_prediction)
    precision = precision_score(y_test, y_prediction, average='weighted')
    recall = recall_score(y_test, y_prediction, average='weighted')
    f1 = f1_score(y_test, y_prediction, average='weighted')

    # Now you can format the accuracy value
    print(f"Training Accuracy for {name}: {training_accuracy:.2f}")
    print(f"Test Accuracy: {accuracy:.2f}")
    print(f"Test Precision: {precision:.2f}")
    print(f"Test Recall: {recall:.2f}")
    print(f"Test F1-Score: {f1:.2f}")

    results.append({
        'Classifier': name,
        'Training Accuracy': training_accuracy,
        'Test Accuracy': accuracy,
        'Test Precision': precision,
        'Test Recall': recall,
        'Test F1-Score': f1
    })

    # Confusion matrix
    conf_matrix = confusion_matrix(y_test, y_prediction)
    plt.figure(figsize=(6, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Confusion Matrix for {name}')
    plt.ylabel('Actual Label', fontsize=14)
    plt.xlabel('Predicted Label', fontsize=14)
    plt.show()

    # Classification report
    print(f"\nClassification Report for {name}:\n")
    print(classification_report(y_test, y_prediction))


# Create a DataFrame from the results
results_df = pd.DataFrame(results)
print(results_df)

```
